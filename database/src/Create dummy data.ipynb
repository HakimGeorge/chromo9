```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker#, relationship, backref

from settings import SETTINGS
from models import Gene, Accession, Exon, Protein
```

```python
class Settings:
    def __init__(self, **attrs):
        for key, value in attrs.iteritems():
            if type(value) is dict:
                attrs[key] = Settings(**value)
        self.__dict__.update(attrs)
```

```python
SETTINGS = Settings(**SETTINGS)

if not hasattr(SETTINGS.DB, 'USER'):
    SETTINGS.DB.USER = raw_input('Please enter username for database : ')

if not hasattr(SETTINGS.DB, 'PASSWD'):
    SETTINGS.DB.PASSWD = raw_input('Enter passwd for mysql user "{}": '.format(SETTINGS.DB.USER))

conn_string = 'mysql+mysqldb://{user}:{passwd}@{location}:{port}'.format(
                                            user = SETTINGS.DB.USER,
                                            passwd = SETTINGS.DB.PASSWD,
                                            location = SETTINGS.DB.LOCATION,
                                            port = SETTINGS.DB.PORT,
                                                                         )
engine = create_engine(conn_string, echo=False)

engine.execute('use {}'.format(SETTINGS.DB.DATABASE_NAME))

Session = sessionmaker(bind=engine)
session = Session()
```

```python
idx = pd.Index(['gene_id','accession_numbers', 'protein_name','protein_sequence','chromosome_location'])
dummy_data = []
for g in session.query(Gene).all()[:100]:
    if len(g.protein_products) > 0:
        for p in g.protein_products:
            row = []
            row.append(g.gene_identifier)
            row.append(':'.join([a.accession_number for a in g.accession_numbers]))
            row.append(p.name)
            row.append(p.sequence)
            row.append(g.chromosome_location)
            dummy_data.append(row)
    else:
        row = []
        row.append(g.gene_identifier)
        row.append(':'.join([a.accession_number for a in g.accession_numbers]))
        row.append('')
        row.append('')
        row.append(g.chromosome_location)
        dummy_data.append(row)
dummy_data = pd.DataFrame(dummy_data)
dummy_data.columns = idx
dummy_data.to_csv('dummy.csv')
dummy_data
```
